### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').mean()['age']
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
occup = users.groupby(['occupation','gender']).gender.count().reset_index(name = 'counts')
occup.head(6)
```

```python
occup['sum1'] = occup.groupby('occupation')['counts'].transform('sum')
```

```python
occup['ratio'] = occup['counts'] / occup['sum1']
#occup[occup['gender']=='F'].sort_values('ratio',ascending = False)
```

```python
occup[occup.gender == 'M'].sort_values('ratio',ascending=False)
```

## 测试使用apply lambda的方式来求百分比

```python
users.head()
```

```python
users.set_index('user_id').head()
```

```python
occup_group = users.groupby(['occupation','gender']).agg(gender_count=('gender','count'))
```

```python
occup_group.head()
```

```python
occup_group.groupby(level=0).apply(lambda x: x / x.sum())
```
